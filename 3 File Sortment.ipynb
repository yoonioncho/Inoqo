```python
import numpy as np
import pandas as pd
```

```python
fgp_analysis=pd.read_csv('[Overview] Final Generic Product Analysis.csv')
fgp_analysis_wo_outlier=pd.read_csv('[Overview] With Outliers Final Generic Product Analysis.csv')
fgp_analysis_wo_outlier=fgp_analysis_wo_outlier.drop(columns=['Unnamed: 0'])
outliers=pd.read_csv('Outliers.csv')
result=pd.read_csv('[RES] Mean Median ML with Deviations.csv')
result=result.drop(columns=['Unnamed: 0','Unnamed: 0.1'])
```

```python
fgp_analysis_wo_outlier=fgp_analysis_wo_outlier.rename(columns={'Outlier Energy':'Outlier energy_per_100g','Without Outlier Energy':'W/O Outlier energy_per_100g'})
fgp_analysis_wo_outlier
```

```python
result_temp=result.drop(columns=['article_number','Product name'])
result_temp=result_temp.drop_duplicates(subset=['Final Generic Product'])
result_temp
```

```python
res=fgp_analysis_wo_outlier.merge(result_temp,how='outer',on=['Final Generic Product'])
```

```python
res=res.drop(columns=['energy_per_100g_y', 'W/O Outlier Mean_y', 'W/O Outlier Median_y'])
```

```python
res=res.rename(columns={'energy_per_100g_x':'energy_per_100g',
                      'W/O Outlier Median_x':'W/O Outlier Median',
       'W/O Outlier Mean_x':'W/O Outlier Mean'})
```

```python
res=res[['Final Generic Product', 'energy_per_100g', 'Number of Products',
       'Min', 'Max', 'Median', 'Mean', 'Range', 'Std',
       'Outlier energy_per_100g', 'W/O Outlier energy_per_100g',
       'W/O Outlier Min', 'W/O Outlier Max', 'W/O Outlier Range', 'W/O Outlier Std',
         'W/O Outlier Median','W/O Outlier Median Deviation',
'W/O Outlier Mean', 'W/O Outlier Mean Deviation',        
       'W/O Outlier ML', 'W/O Outlier ML Deviation']]
```

```python
res.to_csv('[FINAL] Final Generic Product Analysis.csv')
```

```python
result=result[['article_number','Product name','energy_per_100g',
              'Final Generic Product',
              'W/O Outlier Median','W/O Outlier Median Deviation',
'W/O Outlier Mean', 'W/O Outlier Mean Deviation',        
       'W/O Outlier ML', 'W/O Outlier ML Deviation']]
```

```python
result.to_csv('[FINAL] Product Name Analysis for Import.csv')
```

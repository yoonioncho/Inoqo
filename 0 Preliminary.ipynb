```python
# Getting each JSONL line as JSON but only with needed information
import pandas as pd
import jsonlines
import numpy as np

prod=[]
gtin=[]
brand=[]

with jsonlines.open('openfoodfacts-products.jsonl') as reader:
    for obj in reader:
        temp_prod=np.nan
        temp_gtin=np.nan
        temp_brand=np.nan
        if "product_name" in obj:
            temp_prod=obj["product_name"]
        if "code" in obj:
            temp_gtin=obj["code"]
        if "brands" in obj:
            temp_brand=obj["brands"]
        prod.append(temp_prod)
        gtin.append(temp_gtin)
        brand.append(temp_brand)
```

```python
# Setting Dataframe
df = pd.DataFrame(columns = ['Brand', 'Name','GTIN'])
list_of_tuples = list(zip(brand,prod,gtin))
df = pd.DataFrame(list_of_tuples,columns = ['Brand', 'Name','GTIN'])
```
